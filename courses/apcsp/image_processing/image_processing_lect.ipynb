```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("flower.jpg")
```

```python
img.shape
```

```python
fig, ax = plt.subplots()
ax.imshow(img)
```

```python
print(img[200,400,:])
```

```python
img_r = np.zeros(img.shape, dtype="uint8")
img_r[:,:,0] = img[:,:,0]
```

```python
fig, ax = plt.subplots()
ax.imshow(img_r)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
for i in range(3):
    temp = np.zeros(img.shape, dtype="uint8")
    temp[:,:,i] = img[:,:,i]
    axs[i].imshow(temp)
```
