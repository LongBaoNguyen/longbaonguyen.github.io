## Learning goals

````python

# %% [code] Setup
import numpy as np
import matplotlib.pyplot as plt
from math import ceil

from skimage import data, color, transform, exposure, io

np.random.seed(0)
plt.rcParams['figure.figsize'] = (6,4)

def show_grid(images, titles=None, ncols=6, cmap='gray', vmin=0, vmax=1, suptitle=None):
    n = len(images)
    nrows = int(ceil(n / ncols))
    fig, axes = plt.subplots(nrows, ncols, figsize=(2*ncols, 2*nrows))
    axes = np.atleast_1d(axes).ravel()
    for i, ax in enumerate(axes):
        ax.axis('off')
        if i < n:
            ax.imshow(images[i], cmap=cmap, vmin=vmin, vmax=vmax)
            if titles is not None and i < len(titles):
                ax.set_title(titles[i], fontsize=8)
    if suptitle:
        fig.suptitle(suptitle)
    plt.tight_layout(); plt.show()


def to_gray01(img):
    if img.ndim == 3:
        img = color.rgb2gray(img)
    img = img.astype(float)
    if img.max() > 1:
        img = img/255.0
    return img
````


## 1) Small offline dataset (works anywhere)
We synthesize a mini face library from two face‑like images (`camera`, `astronaut`) by applying small rotations, shifts, brightness, gamma, and noise variations—emulating pose/lighting. This avoids downloads in class.


````python

# %% [code] Build a mini face library by augmentation
base1 = to_gray01(data.camera())
base2 = to_gray01(data.astronaut())

H, W = 128, 96
base1r = transform.resize(base1, (H, W), anti_aliasing=True)
base2r = transform.resize(base2, (H, W), anti_aliasing=True)

rots = np.linspace(-10, 10, 5)
shifts = [(0,0), (2,0), (-2,0), (0,2), (0,-2)]
bright_scales = [0.9, 1.0, 1.1]
gammas = [0.9, 1.1]

faces, labels = [], []
for sid, base in enumerate([base1r, base2r]):
    for th in rots:
        rot = transform.rotate(base, th, mode='edge')
        for dy, dx in shifts:
            shifted = np.roll(np.roll(rot, dy, axis=0), dx, axis=1)
            for s in bright_scales:
                img = np.clip(shifted*s, 0, 1)
                for g in gammas:
                    img2 = exposure.adjust_gamma(img, g)
                    img2 = np.clip(img2 + 0.01*np.random.randn(*img2.shape), 0, 1)
                    faces.append(img2)
                    labels.append(sid)

faces = np.array(faces)
labels = np.array(labels)
print('Training faces:', faces.shape, '(n, H, W)')
show_grid(faces[:24], ncols=8, suptitle='Sample training images')
````


## 2) PCA via economy SVD
Vectorize to columns, center by mean face \(\mu\), compute \(X_c=U\Sigma V^T\). Columns of **\(U\)** are the eigenfaces.


````python

# %% [code] Compute SVD and inspect eigenfaces
n, H, W = faces.shape
X = faces.reshape(n, -1).T     # d × n
mu = X.mean(axis=1, keepdims=True)
Xc = X - mu
U, S, Vt = np.linalg.svd(Xc, full_matrices=False)

mean_face = mu.reshape(H, W)
first = [mean_face] + [U[:,i].reshape(H,W) for i in range(12)]
show_grid(first, ncols=7, suptitle='Mean face (left) and first 12 eigenfaces')

explained = (S**2)/np.sum(S**2)
plt.figure(); plt.semilogy(np.arange(1,len(S)+1), S, 'o-')
plt.title('Singular values (log)'); plt.xlabel('k'); plt.ylabel('σ_k'); plt.show()
plt.figure(); plt.plot(np.cumsum(explained), 'o-'); plt.axhline(0.95, color='r', ls='--')
plt.title('Cumulative variance explained'); plt.xlabel('k'); plt.ylabel('cumulative'); plt.show()
````


## 3) Reconstruct a face with top \(r\)
For any image \(x\), \(\hat x = \mu + U_r U_r^T (x-\mu)\).


````python

# %% [code] Reconstruction helper + demo

def reconstruct(img2d, U, mu, r):
    x = img2d.reshape(-1,1)
    a = U[:,:r].T @ (x-mu)
    xhat = mu + U[:,:r] @ a
    return xhat.reshape(img2d.shape)

# Demo with a face-like image
face_img = transform.resize(to_gray01(data.astronaut()), (H, W))
rs = [5,10,25,50,100,200]
recons = [face_img] + [reconstruct(face_img, U, mu, r) for r in rs]
show_grid(recons, titles=['Original']+[f'r={r}' for r in rs], ncols=7,
          suptitle='Face reconstruction with top r eigenfaces')
````


## 4) Cross‑domain reconstruction (coffee/dog)
Approximate a **non‑face** using the face basis. Expect broad, smooth structure capture and missing textures—consistent with the eigenfaces subspace being tuned for faces.


````python

# %% [code] Coffee reconstruction
coffee = transform.resize(to_gray01(data.coffee()), (H, W))
rs = [5,10,25,50,100,200,400]
recons = [coffee] + [reconstruct(coffee, U, mu, r) for r in rs]
show_grid(recons, titles=['Original']+[f'r={r}' for r in rs], ncols=4,
          suptitle='Approximating coffee with eigenfaces')
````


## 5) Try your own image
Upload `dog.jpg` (or any image) next to the notebook, then set `filename` below.


````python

# %% [code] Your image here
filename = 'dog.jpg'  # change to your file
try:
    img = io.imread(filename)
    img = to_gray01(img)
    img = transform.resize(img, (H, W), anti_aliasing=True)
    rs = [5,10,25,50,100,200,400,800]
    recons = [img] + [reconstruct(img, U, mu, r) for r in rs if r <= U.shape[1]]
    show_grid(recons, titles=['Original']+[f'r={r}' for r in rs if r<=U.shape[1]], ncols=4,
              suptitle=f'Approximating {filename} with eigenfaces')
except FileNotFoundError:
    print(f"File '{filename}' not found. Upload and re-run the cell.")
````


## 6) Bonus: 2D projection scatter
Project all training images onto PC1–PC2 and visualize clustering by subject.


````python

# %% [code] Scatter of scores on PC1–PC2
scores = (U[:,:2].T @ (X - mu))   # 2 × n
mask0 = (labels==0); mask1 = (labels==1)
plt.figure(figsize=(5,4))
plt.scatter(scores[0,mask0], scores[1,mask0], s=12, c='tab:blue', label='Subject 0')
plt.scatter(scores[0,mask1], scores[1,mask1], s=12, c='tab:orange', label='Subject 1')
plt.axhline(0,color='k',lw=0.5); plt.axvline(0,color='k',lw=0.5)
plt.legend(); plt.grid(True, ls=':'); plt.title('PC1–PC2 scores'); plt.xlabel('PC1'); plt.ylabel('PC2'); plt.show()
````


## 7) **Colab Option A:** Load a bigger local dataset (Yale/ORL‑like)
Expected folder structure:
```
root/
  subject01/*.png  subject02/*.png  ...
```
Set `root` to your Drive path and choose `resize_to`.


````python

# %% [code] Option A — folder loader + economy SVD
import os
from glob import glob

root = 'data/faces'       # e.g., '/content/drive/MyDrive/faces'
resize_to = (192, 168)    # Yale-like size; use (128,96) if speed matters

all_imgs, labels_big = [], []
subject_dirs = sorted([d for d in glob(os.path.join(root,'*')) if os.path.isdir(d)])
for sid, d in enumerate(subject_dirs):
    files = sorted(glob(os.path.join(d,'*.png')) + glob(os.path.join(d,'*.jpg')))
    for fn in files:
        img = io.imread(fn)
        if img.ndim == 3: img = color.rgb2gray(img)
        img = transform.resize(img.astype(float), resize_to, anti_aliasing=True)
        all_imgs.append(img); labels_big.append(sid)

if all_imgs:
    faces_big = np.array(all_imgs); labels_big = np.array(labels_big)
    print('Loaded', faces_big.shape, 'from', len(subject_dirs), 'subjects')
    X_big = faces_big.reshape(len(faces_big), -1).T
    mu_big = X_big.mean(axis=1, keepdims=True)
    Xc_big = X_big - mu_big
    U_big, S_big, Vt_big = np.linalg.svd(Xc_big, full_matrices=False)
    print('SVD shapes:', U_big.shape, S_big.shape, Vt_big.shape)
    expl_big = (S_big**2)/np.sum(S_big**2)
    plt.figure(); plt.plot(np.cumsum(expl_big)); plt.axhline(0.95,color='r',ls='--')
    plt.title('Big dataset: cumulative variance'); plt.xlabel('k'); plt.ylabel('cumulative'); plt.show()
else:
    print('No images found under', root)
````


## 8) **Colab Option B:** IncrementalPCA for very large n
Streams mini‑batches and approximates PCA subspace while using little memory.


````python

# %% [code] Option B — IncrementalPCA
from sklearn.decomposition import IncrementalPCA

if 'faces_big' in globals():
    imgs = faces_big
    H2, W2 = imgs.shape[1:3]
else:
    imgs = faces
    H2, W2 = imgs.shape[1:3]

X_stream = imgs.reshape(len(imgs), -1)
mu_stream = X_stream.mean(axis=0, keepdims=True)
Xc_stream = X_stream - mu_stream

k = 400     # bump higher on Colab if you like
ipca = IncrementalPCA(n_components=k, batch_size=256)
ipca.fit(Xc_stream)
U_stream = ipca.components_.T              # d × k basis (not exactly orthonormal)
expl_stream = ipca.explained_variance_ratio_
print('IncrementalPCA basis:', U_stream.shape)
plt.figure(); plt.plot(np.cumsum(expl_stream)); plt.axhline(0.95,color='r',ls='--')
plt.title('IncrementalPCA variance'); plt.xlabel('k'); plt.ylabel('cumulative'); plt.show()

# quick reconstruction demo
sample = imgs[0]
flat = sample.reshape(1, -1)
a = (flat - mu_stream) @ U_stream
xhat = mu_stream + a @ U_stream.T
plt.figure(figsize=(6,3))
plt.subplot(1,2,1); plt.imshow(sample, cmap='gray'); plt.title('Original'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(xhat.reshape(H2, W2), cmap='gray'); plt.title('IPCA recon'); plt.axis('off')
plt.tight_layout(); plt.show()
````
